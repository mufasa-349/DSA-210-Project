```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
file_path = "../data/cleaned_study_data.csv"
df = pd.read_csv(file_path)

# Correlation analysis
correlation = df.corr()
print("Correlation Matrix:\n", correlation)

# Heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")

# Save before displaying
plt.savefig("../outputs/correlation_heatmap.png")
plt.show()

# Box Plot: Study Hours vs. Lines of Code Written
plt.figure(figsize=(8, 6))
sns.boxplot(x="Hours Studied", y="Lines of Code Written", data=df, palette="Set2")
plt.title("Box Plot of Study Hours vs. Lines of Code Written")
plt.xlabel("Hours Studied")
plt.ylabel("Lines of Code Written")

# Save the plot before displaying it
plt.savefig("../outputs/box_plot.png")
plt.show()


# Add cumulative columns
df["Cumulative Hours"] = df["Hours Studied"].cumsum()
df["Cumulative Lines"] = df["Lines of Code Written"].cumsum()

# Line Plot: Cumulative Study Hours vs. Cumulative Lines of Code Written
plt.figure(figsize=(8, 6))
plt.plot(df["Cumulative Hours"], df["Cumulative Lines"], marker="o", linestyle="-")
plt.title("Cumulative Study Hours vs. Cumulative Lines of Code Written")
plt.xlabel("Cumulative Study Hours")
plt.ylabel("Cumulative Lines of Code Written")

# Save the plot before displaying it
plt.savefig("../outputs/cumulative_line_plot.png")
plt.show()



# Scatter plot: Study Hours vs. Lines of Code Written
plt.scatter(df["Hours Studied"], df["Lines of Code Written"], color="blue")
plt.title("Study Hours vs. Lines of Code Written")
plt.xlabel("Hours Studied")
plt.ylabel("Lines of Code Written")

# Save the plot before displaying it
plt.savefig("../outputs/scatter_plot.png")
plt.show()
```
