```python
# Import libraries
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load cleaned dataset
file_path = "../data/cleaned_study_data.csv"
df = pd.read_csv(file_path)

# Linear Regression Model Visualization
X = df[["Hours Studied"]]
y = df["Lines of Code Written"]

model = LinearRegression()
model.fit(X, y)

# Calculate R^2 score
r2 = r2_score(y, model.predict(X))

# Scatter plot with regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color="blue", label="Data Points")
plt.plot(X, model.predict(X), color="red", label="Regression Line")
plt.title("Regression Line: Study Hours vs. Lines of Code Written")
plt.xlabel("Hours Studied")
plt.ylabel("Lines of Code Written")
plt.legend()

# Add R^2 score as text on the graph
plt.text(0.05, 0.9, f"$R^2$: {r2:.2f}", transform=plt.gca().transAxes, fontsize=12, color="green", bbox=dict(facecolor="white", alpha=0.5))

# Save the graph
outputs_dir = "../outputs"
if not os.path.exists(outputs_dir):
    os.makedirs(outputs_dir)
plt.savefig(os.path.join(outputs_dir, "regression-line.png"))
plt.show()
```
